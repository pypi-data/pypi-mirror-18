```sos
# This SoS step generates test.jpg by calling a R script, with 
# some variables passed by SoS
output: 'test.jpg'
R:
  library(ggplot2) 
  jpeg(${_output!r})
  qplot(Sepal.Length, Petal.Length, data = iris, color = Species)
  dev.off()
```

```sos
%with R
# here we call R directly by using an IR kernel
library(ggplot2)
qplot(Sepal.Length, Petal.Length, data = iris, geom='smooth')
```

```sos
# here is a SoS section with report
sample_names = ['sampleA', 'sampleB']

! There are ${len(sample_names)} samples with names ${sample_names!,}
```

```sos
# or we can use the script form of the report action to generate
# multi-line Markdown report
report:
  ## samples
  There are ${len(sample_names)} samples with names ${sample_names!,}
```
